## Importando as Bibliotecas Necessárias

```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image

```

## Loading dos Áudios

```python
scale_file = r"D:\IC - Prof Rubelmar\Dataset Soldagem\Audio Soldagem Edição 2\Furou\36\03.wav"
```

```python
ipd.Audio(scale_file)
```

```python
# load audio files with librosa
scale, sr = librosa.load(scale_file)
```

## Mel Filter Banks

```python
filter_banks = librosa.filters.mel(n_fft=2048, sr=22050, n_mels=10)
```

```python
filter_banks.shape
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(filter_banks, 
                         sr=sr, 
                         x_axis="linear")
plt.colorbar(format="%+2.f")
plt.show()
```

## Extraindo um Espectograma Mel

```python
mel_spectrogram = librosa.feature.melspectrogram(scale, sr=sr, n_fft=2048, hop_length=512, n_mels=10)
```

```python
mel_spectrogram.shape
```

```python
log_mel_spectrogram = librosa.power_to_db(mel_spectrogram)
```

```python
log_mel_spectrogram.shape
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(log_mel_spectrogram, 
                         x_axis="time",
                         y_axis="mel", 
                         sr=sr)
plt.colorbar(format="%+2.f") # essa parada aqui representa a coluna de cores

#plt.savefig(r'D:\IC - Prof Rubelmar\Dataset Soldagem\Audio Soldagem Edição 2\saved.png')    
plt.show() #sempre vem depois do savefig()

```
